### GET request

```python
path = "/api/events/"
endpoint = f"{base_url}{path}"
```

```python
response = httpx.get(endpoint)
print('ok', response.is_success)

if response.is_success:
    data = response.json()
    print(data)
```

### PUT request

```python
detail_path = "/api/events/1/"
detail_endpoint = f"{base_url}{detail_path}"
```

```python
response = httpx.put(detail_endpoint, data={'description': 'testing update'})
```

```python
print(response.headers)
print('update succeeded', response.is_success)

if response.is_success:
    data = response.json()
    print(
        data,
        type(data),
        data.get('items')
    )
else:
    print(response)
```

### Sending events

```python
event_count = 100

pages = ['/about', '/contact', '/pages', '/pricing']

for i in range (event_count):
    page = random.choice(pages)
    response = httpx.post(endpoint, json={"page": page})
    if response.is_success:
        data = response.json()
        print (data, type(data), data.get('items'))
    else:
        print(response.text)
```
